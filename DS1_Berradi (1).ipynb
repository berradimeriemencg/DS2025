```python


# Charger un dataset en ligne (exemple : ventes d'une entreprise)
url = "https://raw.githubusercontent.com/datasets/gdp/master/data/gdp.csv"
df = pd.read_csv(url)
```

```python
# Statistiques pour une colonne spécifique
colonne = df['Value']  # Adapter selon vos données
```

```python
#calcul de la moyenne
print(f"Moyenne: {colonne.mean():.2f}")
#calcul de la medianne
print(f"Médiane: {colonne.median():.2f}")
#calcul du mode
print(f"Mode: {colonne.mode()[0]:.2f}")
#calcul de écart-type
print(f"Écart-type: {colonne.std():.2f}")
#calcul de la variance
print(f"Variance: {colonne.var():.2f}")
#calcul de l'étendue
print(f"Min: {colonne.min():.2f}")
print(f"Max: {colonne.max():.2f}")
```

```python
# Quartiles
print(f"Q1 (25%): {colonne.quantile(0.25):.2f}")
print(f"Q2 (50% - Médiane): {colonne.quantile(0.50):.2f}")
print(f"Q3 (75%): {colonne.quantile(0.75):.2f}")
print(f"IQR: {colonne.quantile(0.75) - colonne.quantile(0.25):.2f}")
```

```python
# Configuration du style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

# Créer une figure avec plusieurs sous-graphiques
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Histogramme
axes[0, 0].hist(colonne, bins=30, edgecolor='black', alpha=0.7)
axes[0, 0].set_title('Distribution des valeurs', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Valeur')
axes[0, 0].set_ylabel('Fréquence')

# Ajouter la moyenne et la médiane
axes[0, 0].axvline(colonne.mean(), color='red', linestyle='--',
                   label=f'Moyenne: {colonne.mean():.2f}')
axes[0, 0].axvline(colonne.median(), color='green', linestyle='--',
                   label=f'Médiane: {colonne.median():.2f}')
axes[0, 0].legend()

# 2. Boxplot
axes[0, 1].boxplot(colonne.dropna(), vert=True)
axes[0, 1].set_title('Boxplot - Détection des outliers', fontsize=14, fontweight='bold')
axes[0, 1].set_ylabel('Valeur')

# 3. Densité
colonne.plot(kind='density', ax=axes[1, 0])
axes[1, 0].set_title('Courbe de densité', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Valeur')

# 4. QQ-plot (pour tester la normalité)
from scipy import stats
stats.probplot(colonne.dropna(), dist="norm", plot=axes[1, 1])
axes[1, 1].set_title('Q-Q Plot (Test de normalité)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.patches import Rectangle
import seaborn as sns

# Configuration du style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Données PIB 2024 (en milliards de dollars US)
pays_top10 = {
    'États-Unis': 29167.78,
    'Chine': 18532.63,  # estimation
    'Allemagne': 4659.93,
    'Japon': 4110.45,  # estimation
    'Inde': 3890.24,  # estimation
    'Royaume-Uni': 3495.26,  # estimation
    'France': 3174.10,
    'Italie': 2328.15,  # estimation
    'Brésil': 2188.42,
    'Canada': 2140.00  # estimation
}

# Données PIB par habitant 2024 (en dollars US)
pib_par_habitant = {
    'Monaco': 240862,
    'Liechtenstein': 197504,
    'Luxembourg': 135380,
    'Irlande': 112769,
    'Suisse': 105669,
    'Norvège': 94359,
    'États-Unis': 87756,
    'Allemagne': 55480,
    'France': 47400,
    'Royaume-Uni': 51075
}

# Prévisions de croissance 2024 (en %)
croissance_2024 = {
    'Inde': 7.0,
    'Chine': 4.8,
    'Russie': 3.6,
    'États-Unis': 2.8,
    'Brésil': 2.3,
    'Monde': 3.0,
    'France': 1.1,
    'Allemagne': 0.0,
    'Royaume-Uni': 1.1,
    'Japon': 0.9
}

# Création de la figure avec plusieurs sous-graphiques
fig = plt.figure(figsize=(18, 14))
fig.suptitle('Analyse Complète du PIB Mondial 2024', fontsize=20, fontweight='bold', y=0.995)

# 1. Classement des 10 plus grands PIB mondiaux
ax1 = plt.subplot(2, 3, 1)
pays = list(pays_top10.keys())
valeurs = list(pays_top10.values())
colors = plt.cm.viridis(np.linspace(0.3, 0.9, len(pays)))

bars = ax1.barh(pays, valeurs, color=colors, edgecolor='black', linewidth=1.2)
ax1.set_xlabel('PIB (milliards $US)', fontsize=11, fontweight='bold')
ax1.set_title('Top 10 des PIB Mondiaux 2024', fontsize=13, fontweight='bold', pad=10)
ax1.invert_yaxis()
ax1.grid(axis='x', alpha=0.3, linestyle='--')

# Ajout des valeurs sur les barres
for i, (bar, val) in enumerate(zip(bars, valeurs)):
    ax1.text(val + 500, bar.get_y() + bar.get_height()/2,
             f'{val:,.0f}', va='center', fontsize=9, fontweight='bold')

# 2. Répartition du PIB mondial (Diagramme circulaire)
ax2 = plt.subplot(2, 3, 2)
pib_total = sum(valeurs)
autres = 111326.37 - pib_total  # PIB mondial total - top 10

labels_pie = pays[:5] + ['Reste du monde']
valeurs_pie = valeurs[:5] + [autres + sum(valeurs[5:])]
colors_pie = plt.cm.Set3(np.linspace(0, 1, len(labels_pie)))

wedges, texts, autotexts = ax2.pie(valeurs_pie, labels=labels_pie, autopct='%1.1f%%',
                                     startangle=90, colors=colors_pie,
                                     textprops={'fontsize': 9, 'fontweight': 'bold'})
ax2.set_title('Répartition du PIB Mondial 2024\n(Total: 111 326 Mds $)',
              fontsize=13, fontweight='bold', pad=10)

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

# 3. Taux de croissance prévus 2024
ax3 = plt.subplot(2, 3, 3)
pays_croissance = list(croissance_2024.keys())
taux = list(croissance_2024.values())
colors_growth = ['#2ecc71' if t > 2 else '#f39c12' if t > 1 else '#e74c3c' for t in taux]

bars_growth = ax3.bar(range(len(pays_croissance)), taux, color=colors_growth,
                       edgecolor='black', linewidth=1.2, alpha=0.8)
ax3.set_xticks(range(len(pays_croissance)))
ax3.set_xticklabels(pays_croissance, rotation=45, ha='right', fontsize=9)
ax3.set_ylabel('Taux de croissance (%)', fontsize=11, fontweight='bold')
ax3.set_title('Prévisions de Croissance du PIB 2024', fontsize=13, fontweight='bold', pad=10)
ax3.axhline(y=3.0, color='blue', linestyle='--', linewidth=2, label='Moyenne mondiale (3%)')
ax3.legend(fontsize=9)
ax3.grid(axis='y', alpha=0.3, linestyle='--')

# Ajout des valeurs sur les barres
for i, (bar, val) in enumerate(zip(bars_growth, taux)):
    ax3.text(bar.get_x() + bar.get_width()/2, val + 0.2,
             f'{val:.1f}%', ha='center', fontsize=9, fontweight='bold')

# 4. PIB par habitant (Top 10)
ax4 = plt.subplot(2, 3, 4)
pays_hab = list(pib_par_habitant.keys())
valeurs_hab = list(pib_par_habitant.values())
colors_hab = plt.cm.plasma(np.linspace(0.2, 0.9, len(pays_hab)))

bars_hab = ax4.bar(range(len(pays_hab)), valeurs_hab, color=colors_hab,
                    edgecolor='black', linewidth=1.2)
ax4.set_xticks(range(len(pays_hab)))
ax4.set_xticklabels(pays_hab, rotation=45, ha='right', fontsize=9)
ax4.set_ylabel('PIB/habitant ($US)', fontsize=11, fontweight='bold')
ax4.set_title('Top 10 PIB par Habitant 2024', fontsize=13, fontweight='bold', pad=10)
ax4.grid(axis='y', alpha=0.3, linestyle='--')

# Ajout des valeurs
for i, (bar, val) in enumerate(zip(bars_hab, valeurs_hab)):
    ax4.text(bar.get_x() + bar.get_width()/2, val + 5000,
             f'{val:,.0f}', ha='center', fontsize=8, fontweight='bold', rotation=0)

# 5. Comparaison PIB nominal vs Croissance
ax5 = plt.subplot(2, 3, 5)
pays_compare = ['États-Unis', 'Chine', 'Allemagne', 'France', 'Inde', 'Brésil']
pib_compare = [pays_top10[p] for p in pays_compare]
croissance_compare = [croissance_2024.get(p, 0) for p in pays_compare]

# Normalisation pour la visualisation
pib_norm = [p/max(pib_compare)*100 for p in pib_compare]
x = np.arange(len(pays_compare))
width = 0.35

bars1 = ax5.bar(x - width/2, pib_norm, width, label='PIB (normalisé)',
                color='steelblue', edgecolor='black', linewidth=1)
bars2 = ax5.bar(x + width/2, croissance_compare, width, label='Croissance 2024 (%)',
                color='coral', edgecolor='black', linewidth=1)

ax5.set_xlabel('Pays', fontsize=11, fontweight='bold')
ax5.set_ylabel('Valeur', fontsize=11, fontweight='bold')
ax5.set_title('PIB vs Croissance - Comparaison 2024', fontsize=13, fontweight='bold', pad=10)
ax5.set_xticks(x)
ax5.set_xticklabels(pays_compare, rotation=45, ha='right', fontsize=9)
ax5.legend(fontsize=9)
ax5.grid(axis='y', alpha=0.3, linestyle='--')

# 6. Évolution et statistiques globales
ax6 = plt.subplot(2, 3, 6)
ax6.axis('off')

# Statistiques clés
stats_text = f"""
STATISTIQUES CLÉS DU PIB MONDIAL 2024

📊 PIB Mondial Total: 111 326 milliards $US

🥇 Top 3 Économies:
   1. États-Unis: 29 168 Mds $ (26.2%)
   2. Chine: ~18 533 Mds $ (16.6%)
   3. Allemagne: 4 660 Mds $ (4.2%)

🌍 Croissance Mondiale: 3.0% (prévision)

📈 Croissance la plus forte: Inde (7.0%)

💰 PIB/hab. le plus élevé: Monaco (240 862 $)

🇫🇷 France:
   • Classement mondial: 7ème
   • PIB: 3 174 Mds $
   • PIB/habitant: 47 400 $ (24ème mondial)
   • Croissance 2024: 1.1%

⚠️ Défis 2024:
   • Tensions géopolitiques
   • Inflation persistante
   • Ralentissement immobilier (Chine)
   • Volatilité des marchés
"""

ax6.text(0.1, 0.95, stats_text, transform=ax6.transAxes,
         fontsize=11, verticalalignment='top', family='monospace',
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.3))

plt.tight_layout()
plt.savefig('analyse_pib_2024.png', dpi=300, bbox_inches='tight')
plt.show()

print("=" * 70)
print("ANALYSE DU PIB MONDIAL 2024 - RAPPORT DÉTAILLÉ")
print("=" * 70)
print(f"\n📊 PIB MONDIAL: {111326.37:,.2f} milliards de dollars US\n")

print("🏆 TOP 10 DES ÉCONOMIES MONDIALES:")
print("-" * 70)
for i, (pays, pib) in enumerate(pays_top10.items(), 1):
    part = (pib / 111326.37) * 100
    print(f"{i:2d}. {pays:15s} | PIB: {pib:>10,.2f} Mds $ | Part mondiale: {part:5.2f}%")

print(f"\n💎 TOP 5 PIB PAR HABITANT:")
print("-" * 70)
for i, (pays, pib_hab) in enumerate(list(pib_par_habitant.items())[:5], 1):
    print(f"{i}. {pays:15s} | {pib_hab:>10,.0f} $/habitant")

print(f"\n📈 PRÉVISIONS DE CROISSANCE 2024:")
print("-" * 70)
for pays, taux in sorted(croissance_2024.items(), key=lambda x: x[1], reverse=True):
    emoji = "🚀" if taux >= 5 else "📈" if taux >= 2 else "📊"
    print(f"{emoji} {pays:15s} | {taux:>5.1f}%")

print("\n" + "=" * 70)
print("Graphiques sauvegardés dans 'analyse_pib_2024.png'")
print("=" * 70)
```
